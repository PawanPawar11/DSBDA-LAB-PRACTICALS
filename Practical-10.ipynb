```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
print(iris.head())
```

```python
print("\nFeature Types:")
print(iris.dtypes)
```

```python
iris.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
for i, column in enumerate(iris.select_dtypes(include='number').columns):
    plt.subplot(2, 2, i+1)  # 2 rows, 2 columns layout
    sns.boxplot(y=iris[column])
    plt.title(f"Box plot of {column}")

plt.tight_layout()
plt.show()
```
