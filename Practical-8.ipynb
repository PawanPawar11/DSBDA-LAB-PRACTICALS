```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic, palette='viridis')
```

```python
plt.title('Survival Count of Passengers by their Class')
```

```python
plt.subplot(1, 2, 2)
```

```python
sns.boxplot(x='Survived', y='Age', hue='Survived', data=titanic, palette='magma')
```

```python
plt.title("Age distribution by Survival Status")
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Fare', data=titanic, bins=100, kde=True, color='teal')
plt.title('Distribution of Ticket Price (Fare)')
plt.xlabel("Fare (USD)")
plt.ylabel("Number of Passengers")
plt.xlim(0, 300)
plt.show()
```
